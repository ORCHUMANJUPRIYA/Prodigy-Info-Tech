# Data visualization

```python
import seaborn as sns
```

```python
data['y'].replace({'no':0,'yes':1},inplace=True)
ax = sns.countplot(x='y', data=data, palette='colorblind')
plt.title("Target Distribution")
total = data['y'].count()
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')
plt.show()
```

```python
ax=sns.countplot(x='month',data=data,palette='colorblind')
total=data['month'].count()
plt.title("'month' distribution")
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')
plt.show()
```

```python
data['month'].unique()
```

```python
sns.set_style('whitegrid')
ax=sns.countplot(x='marital',data=data)
plt.title("sex Distribution")
total=data['month'].count()
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')

plt.show()
```

```python
sns.set_style("whitegrid")
ax=sns.countplot(x='education',data=data)
sns.color_palette("cubehelix", as_cmap=True)
total = data['education'].count()

# annotate the bars with fmt from matplotlib v3.7.0
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')

plt.title("'Education' Distribution")
plt.show()
```

```python
cor=data.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

```python
sns.distplot(data['age'],bins=40)#distribution plot
sns.color_palette("Blues", as_cmap=True)
```

```python
sns.set_style("whitegrid")
ax=sns.countplot(x='job',data=data)
total=data['job'].count()
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')
plt.title("'job'Distribution")
plt.xticks(rotation=90)
plt.show()
```
